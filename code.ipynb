# Data Extraction

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Reading the input file
df=pd.read_excel("C:/Users/suraj/Desktop/BlackCoffer/Input.xlsx")

# Iterating over the links and scraping the contents
for index, row in df.iterrows():
    url = row['URL']
    print(url)
    fileName = str(row['URL_ID'])
    print(fileName)
    # Creating new text files to store the scraped data with URL_ID as file name
    file = "C:/Users/suraj/Desktop/BlackCoffer/files/"+fileName+".txt"
    f = open(file, "w",encoding="utf-8")
    page=requests.get(url)
    soup = BeautifulSoup(page.content,'html.parser')
    if(soup.find_all('h1', class_='entry-title')):
        # Scraping the titles
        title = soup.find_all('h1', class_='entry-title')[0]
        f.write(title.string)
    if(soup.find_all('div', class_='td-post-content tagdiv-type')):
        article = soup.find_all('div', class_='td-post-content tagdiv-type')[0]
        contents = article.find_all('p')
        # Scraping the contents
        for para in contents:
            if (para.string!=None):
                f.write(para.string)
    f.close()
```

# NLP and Analysis

```python
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import pandas as pd
import string
import glob
import re
```

## Stop Words Dictionary

```python
# Combining and storing all stop words text files to a single stop words list for ease of processing
read_files = glob.glob("C:/Users/suraj/Desktop/BlackCoffer/StopWords/*.txt")
print(len(read_files))
stop_words = []
for f in read_files:
    with open(f) as stop_text:
        data =stop_text.read().rstrip()
        cleaned_stop_text = data.translate(str.maketrans('', '', string.punctuation))
        tokenized_stop_words = word_tokenize(cleaned_stop_text, "english")
        for word in tokenized_stop_words:
            stop_words.append(word)
print(len(stop_words))
```

## Defining Output Data Variables

```python
Positive_Score = []
Negative_Score = []
Polarity_Score = []
Subjectivity_Score = []
Avg_Sentence_Length = []
Percentage_of_Complex_Words = []
Fog_Index = []
Avg_Number_of_Words_per_Sentence = []
Complex_Word_Count = []
Word_Count = []
Syllable_per_Word = []
Personal_Pronouns = []
Avg_Word_Length = []
```

## Data Analysis

```python
df=pd.read_excel("C:/Users/suraj/Desktop/BlackCoffer/Input.xlsx")

for index, row in df.iterrows():

    fileName = str(row['URL_ID'])
    text = open("C:/Users/suraj/Desktop/BlackCoffer/files/"+fileName+".txt", encoding="utf-8").read()
    
    # Removing punctuations
    cleaned_text = text.translate(str.maketrans('', '', string.punctuation))

    # Tokenizing the cleaned text using NLTK
    tokenized_words = word_tokenize(cleaned_text, "english")

    #
    pronounRegex = re.compile(r'\b(I|we|my|ours|(?-i:us))\b',re.I)
    pronouns = pronounRegex.findall(text)
    # the value obtained from each file is added to the corresponding output list
    Personal_Pronouns.append(len(pronouns))
    
    # Removing stop words from the tokenized words list
    final_words = []
    for word in tokenized_words:
        if word not in stop_words:
            final_words.append(word)
    
    # Positive Score
    Pstv_Txt = 0
    with open("C:/Users/suraj/Desktop/BlackCoffer/Dictionary/positive-words.txt") as positive_text:
        positive_list = word_tokenize(positive_text.read().rstrip(), "english")
        for word in final_words:
            if word.lower() in positive_list:
                Pstv_Txt+=1
        Positive_Score.append(Pstv_Txt)
        
    # Negative Score
    Ngtv_Txt = 0
    with open("C:/Users/suraj/Desktop/BlackCoffer/Dictionary/negative-words.txt") as negative_text:
        negative_list = word_tokenize(negative_text.read().rstrip(), "english")
        for word in final_words:
            if word.lower() in negative_list:
                Ngtv_Txt+=1
        Negative_Score.append(Ngtv_Txt)
        
    # Polarity Score
    Plrty_Scr = (Pstv_Txt-Ngtv_Txt) / ((Pstv_Txt + Ngtv_Txt)+0.000001)
    Polarity_Score.append(Plrty_Scr)
    
    #Subjectivity Score = (Positive Score + Negative Score)/ ((Total Words after cleaning) + 0.000001)
    Sbjctvty_Scr = (Pstv_Txt + Ngtv_Txt) / ((len(final_words))+0.000001)
    Subjectivity_Score.append(Sbjctvty_Scr)
    
    # Removing Stop Words using NLTK for Word Count
    final_words = []
    for word in tokenized_words:
        if word not in stopwords.words('english'):
            final_words.append(word)
    Wrd_Cnt = len(final_words)
    Word_Count.append((Wrd_Cnt))
    
    # Average Word Length
    characters = 0
    characters = sum(len(word) for word in text)
    if(Wrd_Cnt==0):
        Avg_Wrd_Len = 0
    else:
        Avg_Wrd_Len = characters/Wrd_Cnt
    Avg_Word_Length.append(Avg_Wrd_Len)
    
    # Syllable Per Word and Complex Word Count
    Total_Syllable = 0
    Cmplx_Wrd_Cnt = 0
    for word in final_words:
        vowels = "aeiouy"
        Syllable = 0
        for index in range(1,len(word)):
            if word[index] in vowels:
                Syllable+=1
            if word.endswith ("es") or word.endswith("ed"):
                Syllable -=1
        if Syllable > 2:
            Cmplx_Wrd_Cnt +=1
        Total_Syllable += Syllable
    if(Wrd_Cnt==0):
        Sylbl_per_wrd = 0
    else:
        Sylbl_per_wrd = (Total_Syllable/Wrd_Cnt)
    Syllable_per_Word.append(Sylbl_per_wrd)
    Complex_Word_Count.append(Cmplx_Wrd_Cnt)
    
    # Average Sentence Length and Average Number of Words per Sentence
    sentences = len(sent_tokenize(text))
    if(sentences==0):
        Avg_Sent_Len = 0
    else:
        Avg_Sent_Len = Wrd_Cnt/sentences
    Avg_Sentence_Length.append(Avg_Sent_Len)
    Avg_Number_of_Words_per_Sentence.append(Avg_Sent_Len)
    
    # Percentage of Complex Words
    if(Wrd_Cnt==0):
        Per_Cmplx_Wrd = 0
    else:
        Per_Cmplx_Wrd = (Cmplx_Wrd_Cnt/Wrd_Cnt) * 100
    Percentage_of_Complex_Words.append(Per_Cmplx_Wrd)
    
    # Fog Index
    Fg_Indx = 0.4 * (Avg_Sent_Len + Per_Cmplx_Wrd)
    Fog_Index.append(Fg_Indx)

# The length of each list is printed to ensure it contains the data of all the files
print(len(Positive_Score))
print(len(Negative_Score))
print(len(Polarity_Score))
print(len(Subjectivity_Score))
print(len(Avg_Sentence_Length))
print(len(Percentage_of_Complex_Words))
print(len(Fog_Index))
print(len(Avg_Number_of_Words_per_Sentence))
print(len(Complex_Word_Count))
print(len(Word_Count))
print(len(Syllable_per_Word))
print(len(Personal_Pronouns))
print(len(Avg_Word_Length))

# Saving the output list as dataframe columns
df['POSITIVE SCORE'] = Positive_Score
df['NEGATIVE SCORE'] = Negative_Score
df['POLARITY SCORE'] = Polarity_Score
df['SUBJECTIVITY SCORE'] = Subjectivity_Score
df['AVERAGE SENTENCE LENGTH'] = Avg_Sentence_Length
df['PERCENTAGE OF COMPLEX WORDS'] = Percentage_of_Complex_Words
df['FOG INDEX'] = Fog_Index
df['AVERAGE NUMBER OF WORDS PER SENTENCE'] = Avg_Number_of_Words_per_Sentence
df['COMPLEX WORD COUNT'] = Complex_Word_Count
df['WORD COUNT'] = Word_Count
df['SYLLABLE PER WORD'] = Syllable_per_Word
df['PERSONAL PRONOUNS'] = Personal_Pronouns
df['AVERAGE WORD LENGTH'] = Avg_Word_Length

# Saving the data frame to the output file
df.to_excel("C:/Users/suraj/Desktop/BlackCoffer/Output.xlsx", index = False)
df.head()
```
